# 알파벳 찾기

알파벳 소문자로만 이루어진 단어 S가 주어진다. 각각의 알파벳에 대해서, 단어에 포함되어 있는 경우에는 처음 등장하는 위치를, 포함되어 있지 않은 경우에는 -1을 출력하는 프로그램을 작성하시오.

```python
import string

S = list(input())
Alphabet = list(string.ascii_lowercase)
ans = []

for i in range(0,len(Alphabet)):
    if Alphabet[i] in S:
        for j in range(0,len(S)):
            if Alphabet[i] == S[j]:
                ans.append(j)
                break
    else:
        ans.append(-1)
    print(ans[i], end = ' ')
```
