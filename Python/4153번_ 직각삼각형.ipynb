# 직각삼각형

과거 이집트인들은 각 변들의 길이가 3, 4, 5인 삼각형이 직각 삼각형인것을 알아냈다. 주어진 세변의 길이로 삼각형이 직각인지 아닌지 구분하시오.

```python
T = 0

while T == 0:
  lst = []
  a, b,c = map(int, input().split())
  lst.append(a)
  lst.append(b)
  lst.append(c)
  lst.sort()

  if lst[0] == 0 and lst[1] == 0 and lst[2] == 0:
    T += 1
    break
  elif lst[0]**2 + lst[1]**2 == lst[2]**2 :
    print("right")
  else:
    print("wrong")
```
