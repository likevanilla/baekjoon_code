# 럭비 클럽

올 골드 럭비 클럽의 회원들은 성인부 또는 청소년부로 분류된다.

나이가 17세보다 많거나, 몸무게가 80kg 이상이면 성인부이다. 그 밖에는 모두 청소년부이다. 클럽 회원들을 올바르게 분류하라.

```python
T = 0

while T != 1:
  name, age, w = input().split()

  age = int(age)
  w = int(w)

  if age > 17 or w >= 80:
    print(name,"Senior")
  elif name == '#' and age == 0 and w == 0:
    T += 1
  else:
    print(name,"Junior")
```
